```python
import sympy
import numpy as np
from sympy.plotting import plot
from sympy import symbols
```

### 1. Определить корни

```python
x = symbols('x')

sympy.solveset(sympy.Eq(-12*x**4*sympy.sin(sympy.cos(x)) - 18*x**3 + 5*x**2 + 10*x - 30, 0), x)
```

### 2. Найти интервалы, на которых функция возрастает

Интервалов возрастания бесконечное множество.

### 3. Найти интервалы, на которых функция убывает

Интервалов убывания бесконечное множество.

### 4. Построить график

```python

plot(-12*x**4*sympy.sin(sympy.cos(x)) - 18*x**3 + 5*x**2 + 10*x - 30, (x, -100, 100))
```

### 6. Определить промежутки, на котором f > 0 (взят диапазон -10 < x < 10 c шагом 1)

```python
x = np.arange(-10, 10, 1)
func = -12*x**4 * (np.sin(np.cos(x))) - 18*x**3 + 5*x**2 + 10*x - 30

pos_list = []
neg_list = []

for i in x:
    res = -12*i**4 * (np.sin(np.cos(i))) - 18*i**3 + 5*i**2 + 10*i - 30
    if res > 0:
        pos_list.append(res)
    else:
        neg_list.append(res)

print(f'f > 0 в интервале: \n{pos_list[-1]} --> {pos_list[0]}\n')
print(pos_list)
```
